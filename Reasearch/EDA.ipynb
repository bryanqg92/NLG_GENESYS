```python
plt.figure(figsize=(8, 6))  # Ajusta el tamaño de la figura según tus preferencias
cor = df[['P','D','L','R']].corr()
sns.heatmap(cor, annot=True, cmap='coolwarm', vmin=-1, vmax=1)  # Utiliza el mapa de colores 'coolwarm'
plt.title('Matriz de correlación')  # Agrega un título a la figura
plt.xticks(rotation=45)  # Rota las etiquetas del eje x para evitar superposiciones
plt.tight_layout()  # Ajusta el espacio entre los elementos de la figura
plt.show()
```
